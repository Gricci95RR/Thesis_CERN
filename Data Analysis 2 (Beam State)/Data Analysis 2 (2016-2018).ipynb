```python
# Analyzing the distribution between the features when they are split
```

```python
inj_dataframe = df.loc[df['beam_state'] == 'I']
ft_dataframe = df.loc[df['beam_state'] == 'FT']
```

```python
sns.kdeplot(inj_dataframe.height, shade=True);
sns.kdeplot(ft_dataframe.height, shade=True, color='red');
plt.legend(loc='upper left', labels=['Injection', 'Flat Top'])
plt.title('Feature: height')
```

```python
sns.kdeplot(inj_dataframe.position, shade=True);
sns.kdeplot(ft_dataframe.position, shade=True, color='red');
plt.legend(loc='upper left', labels=['Injection', 'Flat Top'])
plt.title('Feature: position')
```

```python
sns.kdeplot(inj_dataframe.decay_a, shade=True);
sns.kdeplot(ft_dataframe.decay_a, shade=True, color='red');
plt.legend(loc='upper left', labels=['Injection', 'Flat Top'])
plt.title('Feature: decay_a')
```

```python
sns.kdeplot(inj_dataframe.decay_b, shade=True);
sns.kdeplot(ft_dataframe.decay_b, shade=True, color='red');
plt.legend(loc='upper left', labels=['Injection', 'Flat Top'])
plt.title('Feature: decay_b')
```
