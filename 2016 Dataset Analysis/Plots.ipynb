```python
ax = sns.countplot(x = "Beam State", data = df, palette = "Set1")
ax.set(xlabel='Beam State', ylabel='Count')
plt.show()
```

```python
ax2 = sns.countplot(x = "Beam Type", data = df, palette = "Set2")
ax2.set(xlabel='Beam Type', ylabel='Count')
plt.show()
```

```python
sns.pairplot(df, palette="husl");
```

```python
sns.pairplot(df, palette="husl", corner=True);
```

```python
g = sns.pairplot(df, diag_kind="kde", corner = True)
```

```python
# Plot Correlation Heatmap
plt.figure(figsize=(15, 8))
correlation = df.corr(method="spearman")
heatmap = sns.heatmap(correlation,vmin=-1, vmax=1, annot=True);
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

```python
g = sns.pairplot(df, kind="reg",corner = True)
```

```python
g = sns.PairGrid(df, hue="Beam Type")
g = g.map_diag(plt.hist, histtype="step", linewidth=3)
g = g.map_offdiag(plt.scatter)
g = g.add_legend()
```
