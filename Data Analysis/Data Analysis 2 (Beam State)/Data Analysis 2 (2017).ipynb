### Detection and Removing the Outliers

```python
df_numerical = df.drop(['collimator_type', 'beam_type', 'beam_state','jaw'],axis=1) # numerical dataframe
df_categorical = df.drop(['height','position','decay_a','decay_b','decay_c'],axis=1) # categorical dataframe
```

```python
Q1 = df_numerical.quantile(0.25)
Q3 = df_numerical.quantile(0.75)
IQR = Q3 - Q1
idx = ~((df_numerical < (Q1 - 1.5 * IQR)) | (df_numerical > (Q3 + 1.5 * IQR))).any(axis=1) # index of rows to keep and discard
df_cleaned = pd.concat([df_numerical.loc[idx], df_categorical.loc[idx]], axis=1)
```
