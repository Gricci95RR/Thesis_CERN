**Distributions** between the features when they are splitted

```python
inj_dataframe = df.loc[df['beam_state'] == 'I']
ft_dataframe = df.loc[df['beam_state'] == 'FT']
```

```python
sns.kdeplot(inj_dataframe.height, shade=True);
sns.kdeplot(ft_dataframe.height, shade=True, color='red');
plt.legend(loc='upper left', labels=['Injection', 'Flat Top'])
plt.title('Feature: height')
```

The feature "height" at Flat Top is right skewed/positively skewed. 

```python
sns.kdeplot(inj_dataframe.position, shade=True);
sns.kdeplot(ft_dataframe.position, shade=True, color='red');
plt.legend(loc='upper left', labels=['Injection', 'Flat Top'])
plt.title('Feature: position')
```

The distribution for the feature "position" is about the same and they are right skewed/positively skewed.

```python
sns.kdeplot(inj_dataframe.decay_a, shade=True);
sns.kdeplot(ft_dataframe.decay_a, shade=True, color='red');
plt.legend(loc='upper left', labels=['Injection', 'Flat Top'])
plt.title('Feature: decay_a')
```

At FT the values of the feature "decay_a" are all around 0, while at injection are more distributed in the x-axis

```python
sns.kdeplot(inj_dataframe.decay_b, shade=True);
sns.kdeplot(ft_dataframe.decay_b, shade=True, color='red');
plt.legend(loc='upper left', labels=['Injection', 'Flat Top'])
plt.title('Feature: decay_b')
```

The distribution for the feature "decay_b is about the same and they are  right skewed/positively skewed."

```python
sns.kdeplot(inj_dataframe.decay_c, shade=True);
sns.kdeplot(ft_dataframe.decay_c, shade=True, color='red');
plt.legend(loc='upper left', labels=['Injection', 'Flat Top'])
plt.title('Feature: decay_c')
```

At FT the values of the feature "decay_c" are all around 0, while at injection are distributed in the x-axis

```python
sns.kdeplot(inj_dataframe.jaw, shade=True);
sns.kdeplot(ft_dataframe.jaw, shade=True, color='red');
plt.legend(loc='upper left', labels=['Injection', 'Flat Top'])
plt.title('Feature: jaw')
```

```python
sns.kdeplot(inj_dataframe.spike, shade=True);
sns.kdeplot(ft_dataframe.spike, shade=True, color='red');
plt.legend(loc='upper left', labels=['Injection', 'Flat Top'])
plt.title('Feature: spike')
```

A **box plot** (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be â€œoutliersâ€ using a method that is a function of the inter-quartile range.
